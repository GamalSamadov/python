```python
import pandas as pd

df = pd.read_csv('marks2.csv')

# df.info()

dfstat = pd.DataFrame(columns=['Stat', 'Class1', 'Class2'])

# Count 
x = df['Class1'].count()
y = df['Class2'].count()
row = {
    'Stat':'Count',
    'Class1' : x,
    'Class2' : y,
    }

dfstat = dfstat.append(row, ignore_index=True)


# min 
x = df['Class1'].min()
y = df['Class2'].min()
row = {
    'Stat':'MIN',
    'Class1' : x,
    'Class2' : y,
    }

dfstat = dfstat.append(row, ignore_index=True)


# max 
x = df['Class1'].max()
y = df['Class2'].max()
row = {
    'Stat':'MAX',
    'Class1' : x,
    'Class2' : y,
    }

dfstat = dfstat.append(row, ignore_index=True)


# mean 
x = df['Class1'].mean()
y = df['Class2'].mean()
row = {
    'Stat':'Mean',
    'Class1' : x,
    'Class2' : y,
    }

dfstat = dfstat.append(row, ignore_index=True)


# median 
x = df['Class1'].median()
y = df['Class2'].median()
row = {
    'Stat':'Median',
    'Class1' : x,
    'Class2' : y,
    }

dfstat = dfstat.append(row, ignore_index=True)


# Standard devision 
x = df['Class1'].std()
y = df['Class2'].std()
row = {
    'Stat':'Std',
    'Class1' : x,
    'Class2' : y,
    }

dfstat = dfstat.append(row, ignore_index=True)


# Mode 
x = df['Class1'].mode()[0]
y = df['Class2'].mode()[0]
print(x, y)
row = {
    'Stat':'Mode',
    'Class1' : x,
    'Class2' : y,
    }

dfstat = dfstat.append(row, ignore_index=True)


dfstat
```
