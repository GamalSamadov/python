```python
import pandas as pd

df = pd.read_csv('diabetes.csv')

df.head()
```

```python
df.info()
```

```python
correletions = df.corr(method='pearson')
correletions
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(11,11))

ax = fig.add_subplot(1,1,1)
cax = ax.matshow(correletions, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = [0,1,2,3,4,5,6,7,8]
ax.set_xticks(ticks)
ax.set_yticks(ticks)
labels = ['Pregnancies', 'Glucose', 'BP', 'SkinThickness', 'Insulin', 'BMI', 'DPF', 'Age', 'Outcome']
ax.set_xticklabels(labels)
ax.set_yticklabels(labels)
plt.show()

```

```python
xpoints = df['BMI']
ypoints = df['BloodPressure']
plt.scatter(xpoints, ypoints)
plt.xlabel('BMI')
plt.ylabel('BloodPressure')
plt.show()
```

```python
import seaborn as sns

sns.lineplot(data=df, x='Age', y='BloodPressure')
```
